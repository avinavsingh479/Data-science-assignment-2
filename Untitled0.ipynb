```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast # Used for safely evaluating string literals containing Python structures
```

```python
movies_df = pd.read_csv('imdb_data (1).csv') # Renamed DataFrame for better context
print("First 5 rows of the dataset:")
display(movies_df.head()) # Using display for richer output
```

```python
# Filter out movies with no budget, as they cannot have a meaningful ROI
movies_df = movies_df[movies_df['budget'] > 0]

# Calculate movie profit (revenue - budget)
movies_df['profit'] = movies_df['revenue'] - movies_df['budget']

# Calculate Return On Investment (ROI)
movies_df['ROI'] = movies_df['profit'] / movies_df['budget']

print("DataFrame updated with 'profit' and 'ROI' columns. First 5 rows:")
display(movies_df.head())
```

```python
# Helper function to extract director's name from the crew list
def get_movie_director(crew_json_str):
    try:
        crew_list = ast.literal_eval(crew_json_str)
        for person in crew_list:
            if person.get('job') == 'Director':
                return person.get('name')
    except (ValueError, TypeError):
        return None

# Helper function to extract producer's name from the crew list
def get_movie_producer(crew_json_str):
    try:
        crew_list = ast.literal_eval(crew_json_str)
        for person in crew_list:
            if person.get('job') == 'Producer':
                return person.get('name')
    except (ValueError, TypeError):
        return None

# Helper function to extract actor names from the cast list
def get_movie_actors(cast_json_str):
    try:
        cast_list = ast.literal_eval(cast_json_str)
        return [actor.get('name') for actor in cast_list if actor.get('name')]
    except (ValueError, TypeError):
        return []

# Apply the extraction functions to create new columns
movies_df['director'] = movies_df['crew'].apply(get_movie_director)
movies_df['producer'] = movies_df['crew'].apply(get_movie_producer)
movies_df['actors'] = movies_df['cast'].apply(get_movie_actors)

# Remove rows where no actors could be extracted (empty actor list)
movies_df = movies_df[movies_df['actors'].map(len) > 0]

print("DataFrame updated with 'director', 'producer', and 'actors' columns. First 5 rows after filtering:")
display(movies_df.head())
```

```python
# Find the movie with the highest profit
highest_profit_movie = movies_df.loc[movies_df['profit'].idxmax()]

print("\n--- Movie with the Highest Profit ---")
print(f"Title: {highest_profit_movie['title']}")
print(f"Director: {highest_profit_movie['director']}")
print(f"Producer: {highest_profit_movie['producer']}")
print(f"Top 3 Actors: {highest_profit_movie['actors'][:3]}")
print(f"Total Profit: ${highest_profit_movie['profit']:,}")
```

```python
# Select the top 10 movies by profit
top_10_profit_movies = movies_df.sort_values(by='profit', ascending=False).head(10)

# Create a bar plot for the top 10 movies by profit
plt.figure(figsize=(12, 7)) # Adjust figure size for better readability
sns.barplot(x=top_10_profit_movies['profit'], y=top_10_profit_movies['title'], palette='viridis') # Changed palette
plt.title("Top 10 Movies by Profit (Revenue - Budget)", fontsize=16) # More descriptive title
plt.xlabel("Profit (USD)", fontsize=12)
plt.ylabel("Movie Title", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7) # Add grid for better comparison
plt.tight_layout() # Adjust layout
plt.show()
```

```python
# Group by original language and calculate the mean ROI, then sort
language_roi_mean = movies_df.groupby('original_language')['ROI'].mean().sort_values(ascending=False)

print("\n--- Top 5 Original Languages by Average ROI ---")
print(language_roi_mean.head())
```

```python
# Create a bar plot for the top 10 languages by ROI
plt.figure(figsize=(10, 6))
sns.barplot(x=language_roi_mean.head(10).index, y=language_roi_mean.head(10).values, palette='magma') # Changed palette
plt.xticks(rotation=60, ha='right') # Rotate and align x-axis labels
plt.title("Top 10 Original Languages by Average ROI", fontsize=16)
plt.xlabel("Original Language", fontsize=12)
plt.ylabel("Average ROI", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Get the counts of the top 5 most frequent original languages
top_5_languages_count = movies_df['original_language'].value_counts().head(5)

# Create a pie chart to visualize the distribution of these top languages
plt.figure(figsize=(8, 8))
plt.pie(top_5_languages_count.values, labels=top_5_languages_count.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel')) # Added startangle and custom colors
plt.title("Distribution of Top 5 Original Languages", fontsize=16)
plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
all_unique_genres = set() # Renamed variable

# Iterate through each movie's genres and add them to the set
for genre_str in movies_df['genres'].dropna():
    try:
        parsed_genre_list = ast.literal_eval(genre_str)
        for genre_item in parsed_genre_list:
            if 'name' in genre_item:
                all_unique_genres.add(genre_item['name'])
    except (ValueError, TypeError):
        # Handle cases where the string is not a valid literal
        pass

print("\n--- All Unique Genres Found in the Dataset ---")
print(all_unique_genres)
```

```python
# To represent 'unique genres' as a countplot, we need a list where each unique genre appears once.
# The original code produced a bar for each unique genre as if its count was 1.
# We will replicate that specific visual output.
visual_genre_list = list(all_unique_genres)

plt.figure(figsize=(10, 8))
sns.countplot(y=visual_genre_list, order=visual_genre_list, palette='coolwarm') # Using the list itself as data for countplot, changed palette
plt.title("Listing of Unique Movie Genres", fontsize=16) # Changed title to reflect the actual plot content
plt.xlabel("Count (Each genre listed once)", fontsize=12) # Clarified x-axis label
plt.ylabel("Genre Name", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("\n--- Sample of Movie Titles, Producers, and Directors ---")
display(movies_df[['title','producer','director']].head())

# Group by producer and calculate the average ROI, then get the top 3
top_3_producers_by_roi = movies_df.groupby('producer')['ROI'].mean().sort_values(ascending=False).head(3)

print("\n--- Top 3 Producers by Average ROI ---")
print(top_3_producers_by_roi)
```

```python
# Create a bar plot for the top 3 producers by ROI
plt.figure(figsize=(8, 5))
top_3_producers_by_roi.plot(kind='bar', color=sns.color_palette('rocket')[0]) # Changed color
plt.title("Top 3 Producers by Average ROI", fontsize=16)
plt.xlabel("Producer Name", fontsize=12)
plt.ylabel("Average ROI", fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Create a Series where each actor from every movie is a separate entry
expanded_actors_series = movies_df['actors'].explode()

# Find the actor who appears in the most movies
most_frequent_actor = expanded_actors_series.value_counts().idxmax()
print(f"\n--- Actor Appearing in the Most Movies: {most_frequent_actor} ---")

# Filter the DataFrame to show movies starring the most frequent actor
movies_with_top_actor = movies_df[movies_df['actors'].apply(lambda actor_list: most_frequent_actor in actor_list)]
print(f"\nMovies starring {most_frequent_actor} and their profits:")
display(movies_with_top_actor[['title','profit']])
```

```python
# Create a histogram to visualize the profit distribution of movies starring the most frequent actor
plt.figure(figsize=(10, 6))
sns.histplot(movies_with_top_actor['profit'], kde=True, color='skyblue', bins=20) # Added color and bins for clarity
plt.title(f"Profit Distribution of Movies Starring {most_frequent_actor}", fontsize=16)
plt.xlabel("Profit (USD)", fontsize=12)
plt.ylabel("Number of Movies", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Get the top 3 directors based on the number of movies they directed
top_3_directors_by_movie_count = movies_df['director'].value_counts().head(3).index

print("\n--- Most Frequent Collaborating Actor for Each of the Top 3 Directors ---")
for current_director in top_3_directors_by_movie_count:
    # Filter movies by the current director
    director_specific_movies = movies_df[movies_df['director'] == current_director]
    # Explode the actors list to count individual actor appearances for this director
    actors_for_this_director = director_specific_movies['actors'].explode()

    print(f"\nDirector: {current_director}")
    # Print the most frequent actor for this director
    print(actors_for_this_director.value_counts().head(1))
```

```python
print("\n--- Top Collaborating Actors for Each of the Top 3 Directors (Visualized) ---")
for director_name in top_3_directors_by_movie_count:
    # Filter movies by the current director
    director_subset_df = movies_df[movies_df['director'] == director_name]
    # Get the top 5 most frequent actors for this director
    top_5_actors_for_director = director_subset_df['actors'].explode().value_counts().head(5)

    # Create a bar plot for the top actors of the current director
    plt.figure(figsize=(9, 6))
    sns.barplot(x=top_5_actors_for_director.values, y=top_5_actors_for_director.index, palette='plasma') # Changed palette
    plt.title(f"Top 5 Actors for Director: {director_name}", fontsize=16)
    plt.xlabel("Number of Collaborations", fontsize=12)
    plt.ylabel("Actor Name", fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
